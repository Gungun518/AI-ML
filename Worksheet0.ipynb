```python
#4.1 Exercise on Functions
# Task 1
# Unit Converter Program

def convert_units(choice, value):
    """
    Perform unit conversion based on user's choice.

    Parameters:
        choice (int): Conversion option selected by user
        value (float): Value that needs to be converted

    Returns:
        float or None: Converted value if valid choice,
        otherwise None
    """

    # Length
    if choice == 1:          # meter → feet
        return value * 3.28084
    elif choice == 2:        # feet → meter
        return value / 3.28084

    # Weight
    elif choice == 3:        # kg → pounds
        return value * 2.20462
    elif choice == 4:        # pounds → kg
        return value / 2.20462

    # Volume
    elif choice == 5:        # liter → gallon
        return value * 0.264172
    elif choice == 6:        # gallon → liter
        return value / 0.264172

    else:
        return None


# -------- main program --------
print("Simple Unit Converter")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    user_choice = int(input("Enter your choice (1-6): "))
    user_value = float(input("Enter the value: "))

    answer = convert_units(user_choice, user_value)

    if answer is None:
        print("Invalid conversion type selected.")
    else:
        print("Converted value =", round(answer, 4))

except ValueError:
    print("Invalid input! Please enter numeric values only.")
```

```python

# 1. Extract Every Other Element
def extract_every_other(lst):
    """
    Return every other element starting from the first element.

    Parameters:
        lst (list): Input list.

    Returns:
        list: List containing every alternate element.
    """
    return lst[::2]


# Example
print(extract_every_other([1, 2, 3, 4, 5, 6]))
```

4.2 Exercise on List Manipulation:

```python
#2. Slice a Sublist
def get_sublist(lst, start, end):
    """
    Return a sublist from start to end index (inclusive).

    Parameters:
        lst (list): Input list.
        start (int): Starting index.
        end (int): Ending index (inclusive).

    Returns:
        list: Extracted sublist.
    """
    return lst[start:end + 1]


# Example
print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

```python
# Reverse a List using SLicing
def reverse_list(lst):
    """
    Return the reversed list using slicing.

    Parameters:
        lst (list): Input list.

    Returns:
        list: Reversed list.
    """
    return lst[::-1]


# Example
print(reverse_list([1, 2, 3, 4, 5]))
```

```python
# Remove the first and last elements
def remove_first_last(lst):
    """
    Remove the first and last elements of the list.

    Parameters:
        lst (list): Input list.

    Returns:
        list: List without first and last elements.
    """
    return lst[1:-1]


# Example
print(remove_first_last([1, 2, 3, 4, 5]))
```

```python
# Get the first n elements
def get_first_n(lst, n):
    """
    Return the first n elements of the list.

    Parameters:
        lst (list): Input list.
        n (int): Number of elements.

    Returns:
        list: First n elements.
    """
    return lst[:n]


# Example
print(get_first_n([1, 2, 3, 4, 5], 3))
```

```python
# Extract Last n elements
def get_last_n(lst, n):
    """
    Return the last n elements of the list.

    Parameters:
        lst (list): Input list.
        n (int): Number of elements.

    Returns:
        list: Last n elements.
    """
    return lst[-n:]


# Example
print(get_last_n([1, 2, 3, 4, 5], 2))
```

```python
# Extract Elements in Reverse Skip Order
def reverse_skip(lst):
    """
    Return every second element in reverse order starting
    from the second-to-last element.

    Parameters:
        lst (list): Input list.

    Returns:
        list: Extracted elements.
    """
    return lst[-2::-2]


# Example
print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

4.3 Exercise on Nested List:

```python
def flatten(lst):
    """
    Flatten a nested list by one level.

    Parameters:
        lst (list): Nested list.

    Returns:
        list: Flattened list.
    """
    result = []
    for item in lst:
        result.extend(item)
    return result


# Example
print(flatten([[1, 2], [3, 4], [5]]))
```

```python
def access_nested_element(lst, indices):
    """
    Access an element from a nested list using given indices.

    Parameters:
        lst (list): Nested list.
        indices (list): List of indices.

    Returns:
        element: The accessed element.
    """
    element = lst
    for idx in indices:
        element = element[idx]
    return element


# Example
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(access_nested_element(lst, [1, 2]))
```

```python
def sum_nested(lst):
    """
    Calculate the sum of all numbers in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        int/float: Sum of all elements.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# Example
print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

```python
def remove_element(lst, elem):
    """
    Remove all occurrences of elem from a nested list.

    Parameters:
        lst (list): Nested list.
        elem: Element to remove.

    Returns:
        list: Modified list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            cleaned = remove_element(item, elem)
            if cleaned:
                result.append(cleaned)
        elif item != elem:
            result.append(item)
    return result


# Example
lst = [[1, 2], [3, 2], [4, 5]]
print(remove_element(lst, 2))
```

```python
def find_max(lst):
    """
    Find the maximum element in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        int/float: Maximum element.
    """
    max_val = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val


# Example
print(find_max([[1, 2], [3, [4, 5]], 6]))

```

```python
def count_occurrences(lst, elem):
    """
    Count how many times elem appears in a nested list.

    Parameters:
        lst (list): Nested list.
        elem: Element to count.

    Returns:
        int: Number of occurrences.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


# Example
lst = [[1, 2], [2, 3], [2, 4]]
print(count_occurrences(lst, 2))  # Output: 3
```

```python
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list.

    Parameters:
        lst (list): Deeply nested list.

    Returns:
        list: Fully flattened list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


# Example
print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
# Output: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
def average_nested(lst):
    """
    Calculate the average of all elements in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        float: Average value.
    """
    flat = deep_flatten(lst)
    return sum(flat) / len(flat) if flat else 0


# Example
print(average_nested([[1, 2], [3, 4], [5, 6]]))  # Output: 3.5
```
